```python
#Problem #1
words = input().split(" ")
unique = list(set(words))

print(f"Unique words count: {len(unique)}\nUnique words: {set(unique)}")
```

```python
#Problem #2
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

max_value = max(employees.values())
max_value_name = max(employees, key=employees.get)

print(f"Highest Salary: {max_value_name} - {max_value}")
```

```python
#Problem #3
numbers = [45, 22, 89, 10, 66]

print(f"List: {numbers}\nMax: {max(numbers)}\nMin: {min(numbers)}")
```

```python
#Problem #4
prices = [450, 1200, 899, 1500, 300]

print(f"Products Above 1000: {sum([prices.count(price) for price in prices if price > 1000])}")
```

```python
#Problem #5
attendance = ["P", "P", "A", "P", "P"]

print(f"Attendance Percentage: {(attendance.count("P") / len(attendance)) * 100}")
```

```python
#Problem #6
phone_numbers = [9876543210, 9123456789, 9876543210]

print(f"Unique phone numbers: {set(phone_numbers)}")
```
